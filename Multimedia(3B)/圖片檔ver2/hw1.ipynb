```python
img=io.imread('0413.jpg') 
rows,cols,dims=img.shape

row1 = int(rows/2)
img1 = np.uint8(np.zeros((rows,cols,3))) 
img2 = np.uint8(np.zeros((rows,cols,3))) 
img3 = np.uint8(np.zeros((rows,cols,3)))
img4=img.copy() 
col4 = int(cols/2)

img5=img.copy()

row6 = int(rows/3) 
col6 = int(cols/3)
img6 = np.uint8(np.zeros((row6,col6,3)))

row7 = int(rows*2) 
col7 = int(cols*2)
img7 = np.uint8(np.zeros((row7,col7,3))) 
img8 = np.uint8(np.zeros((cols,rows,3))) 
img9 = np.uint8(np.zeros((rows,cols,3)))
plt.figure(num='P',figsize=(12,12))#1 inch = 0.39cm

#origin 
plt.subplot(3,3,5) 
plt.title('origin image')
plt.imshow(img5)

#half down
for i in range(0,row1):
    for j in range(0,cols):
        for k in range(0,3):
            img1[i,j,k] = img[rows-row1+i,j,k]

for i in range(row1,rows):
    for j in range(0,cols):
        for k in range(0,3): 
            img1[i,j,k] = img[i,j,k]
            
plt.subplot(3,3,1)
plt.title('half down')
plt.imshow(img1)

#mirror updown
for i in range(0,rows):
    for j in range(0,cols):
        for k in range(0,3):
            img2[i,j,k] = img[rows-1-i,j,k] 
plt.subplot(3,3,2)
plt.title('mirror updown') 
plt.imshow(img2)

#mirror leftright
for i in range(0,rows):
    for j in range(0,cols):
        for k in range(0,3): 
            img3[i,j,k] = img[i,cols-1-j,k]
plt.subplot(3,3,3) 
plt.title('mirror leftright') 
plt.imshow(img3)
io.imsave('mirrorleftright.jpg',img3)

#half left
for i in range(0,rows):
        for j in range(0,col4):
            for k in range(0,3):
                img4[i,j,k] = img4[i,j,k]

for i in range(0,rows):
        for j in range(col4,cols):
            for k in range(0,3):
                img4[i,j,k] = img4[i,j-col4,k]
plt.subplot(3,3,4)
plt.title('half left') 
plt.imshow(img4)

#decrease to 1/3
for i in range(0,row6):
    for j in range(0,col6):
        for k in range(0,3):
            sum = 0
            for l in range(0,3):
                for m in range(0,3):
                    sum = sum + img[3*i+l,3*j+m,k]
            img6[i,j,k] = np.uint8(sum/9)
plt.subplot(3,3,6)
plt.title('decrease 1/3')
plt.imshow(img6)

#magnify 4 times
for i in range(0,rows):
    for j in range(0,cols):
        for k in range(0,3):
            for l in range(2*i,2*(i+1)):
                for m in range(2*j,2*(j+1)): 
                    img7[l,m,k] = img[i,j,k]
plt.subplot(3,3,7) 
plt.title('magnify 2 times')
plt.imshow(img7)


for i in range(0,rows):
    for j in range(0,cols):
        for k in range(0,3): 
            img8[j,rows-1-i,k] = img[i,j,k]
plt.subplot(3,3,8)
plt.title('rotate right 90')
plt.imshow(img8)

#rotate 180 degree
for i in range(0,rows):
    for j in range(0,cols):
        for k in range(0,3):
            img9[rows-1-i,cols-1-j,k] = img[i,j,k] 
plt.subplot(3,3,9)
plt.title('rotate 180')
plt.imshow(img9)
plt.show()
```

```python
def distance(color1, color2):
    dif = 0
    for i in range(0,3):
        dif = dif + math.pow((color1[i]-color2[i]),2)
    return math.sqrt(dif)

def makeColor(image,brown): 
    rows,cols,dims=image.shape 
    for i in range(0,rows):
        for j in range(0,cols):
            if distance(image[i,j],brown)<50.0:
                image[i,j,0] = 255
    return image

img=io.imread('ape.jpg') 
io.imshow(img)

brown = [45, 25,15]
img2 = makeColor(img, brown)
plt.figure()
io.imshow(img) 
io.show()
```

```python
from skimage import io
import matplotlib.pyplot as plt
import math
import numpy as np

def distance(color1, color2):
    dif = 0
    for i in range(0,3):
        dif = dif + math.pow((color1[i]-color2[i]),2)
    return math.sqrt(dif)

def chromakey(image,newBg,thresh,bgColor):
    rows,cols,dims=image.shape #print(rows)
    ret_img = np.uint8(np.zeros((rows,cols,3)))
    for i in range(0,rows):
        for j in range(0,cols):
            dif = distance(image[i,j],bgColor)
            if dif < thresh:
                for k in range(0,3):
                    ret_img[i,j,k] = newBg[i,j,k]
            else:
                for k in range(0,3):
                    ret_img[i,j,k] = image[i,j,k]
    return ret_img

img=io.imread('chromakey.jpg')
io.imshow(img)

img2=io.imread('jungle1.jpg')
plt.figure()
io.imshow(img2)

bgcolor = [0, 255, 0]
threshold = 100

img3 = chromakey(img, img2, threshold, bgcolor)
plt.figure()
io.imshow(img3)
io.show()
```

```python
from skimage import io
import matplotlib.pyplot as plt

def edgeDetection(image,thre):
    rows,cols,dims=image.shape 
    for i in range(0,rows-1):
        for j in range(0,cols-1):
            sum = 0 
            sum_rightBottom = 0
            for k in range(0,3):
                sum = sum + image[i,j,k]
                sum_rightBottom = sum_rightBottom + image[i+1,j+1,k]
            dif = abs(sum_rightBottom - sum) 
            if dif > thre:
                value = 0
            else:
                value = 255
            for k in range(0,3):
                image[i,j,k] = value
    return image

img=io.imread('scene.jpg')
threshold = 50

plt.figure(num='scene', figsize=(20,20))

plt.subplot(1,2,1) 
plt.title('origin image') 
plt.imshow(img)

img2 = edgeDetection(img, threshold)

plt.subplot(1,2,2)
plt.title('edge detected image')
plt.imshow(img2)
plt.show()
```
